# Lab | Web Scraping Multiple Pages

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.officialcharts.com/charts/end-of-year-singles-chart/"
```

```python
main > article > div.grid > div.grid__cell.unit-2-3--desktop > section > table > tbody > tr:nth-child(2) > td:nth-child(3) > div > div.title-artist > div.title > a
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
soup.find_all("div", class_ = "artist")[1].get_text()
```

```python
for i in soup.select("div.artist"):
    print(i.a.get_text())
```

```python
title = []
for i in soup.select("div.title"):
    title.append((i.a.get_text()))
```

```python
artist = []
for i in soup.select("div.artist"):
    artist.append((i.a.get_text()))
```

```python
len(title)
```

```python
len(artist)
```

```python
official_chart = pd.DataFrame({'title':title, 'artist': artist})
```

```python
official_chart
```

```python
musicBox = pd.concat([billboard, official_chart], axis = 0)
```
